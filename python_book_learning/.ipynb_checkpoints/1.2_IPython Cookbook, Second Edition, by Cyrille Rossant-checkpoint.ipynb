```python
#url变量包含了csv文件
#pandas定义了一个read_csv()函数可以读csv文件，自动下载和解析该文件并且返回DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
url = ("https://raw.githubusercontent.com/"
       "ipython-books/cookbook-2nd-data/"
       "master/bikes.csv")
df = pd.read_csv(url, index_col='Date',
                parse_dates=True, dayfirst=True)
df.head(2) #df包含DataFrame，包含2维表格式数据，head(2)中2表示展示2行表格
```

```python
#获取表格的一些统计数据
df.describe()
```

```python
#挑选两列Berri1和PierDup，以图表形式查看两条轨迹
df[['Berri1','PierDup']].plot(figsize=(10,6),style=['-','--'],lw=2)
```

```python
#显示df.index下的weekday_name
df.index.weekday_name
```

```python
#把table元素通过weekday分组(Monday is 0, Tuesday is 1, and so on)
df_week = df.groupby(df.index.weekday).sum()
df_week
```

```python
#用plot()创建plot（即图）
fig, ax = plt.subplots(1,1,figsize=(10,8))
df_week.plot(style='-0',lw=3, ax=ax)
ax.set_xlabel('Weekday')
#我们用labels 0, 1 ,2 ...来代替weekday的名字
ax.set_xticklabels(
    ('Monday,Tuesday,Wednesday,Thursday,''Friday,Saturday,Sunday').split(',')
    )
ax.set_ylim(0)
```

```python
from ipywidgets import interact

@interact
def plot(n=(1,30)):
    fig , ax = plt.subplots(1,1, figsize=(10,8))
    df['Berri1'].rolling(window=n).mean().plot(ax=ax)
    plt.show()
```
